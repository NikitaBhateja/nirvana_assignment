```python
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
data1 = pd.read_csv("data_set_1_claim_count_data_2019_2020.csv")
data1.shape
```

```python
data1.head()
```

```python
data1['claim counts'].value_counts()
```

```python
data1['claim_frequency'] = data1['claim counts']/data1['insured years']
data1.claim_frequency.describe()
```

```python
data1.isnull().sum()
```

```python
data1[data1['driver gender'].isnull()]
```

```python
data1[data1['driver gender']=='common goods']
```

```python
print(data1.shape)
data1 = data1[(~data1['driver gender'].isnull())&(data1['driver gender']!='common goods')]
data1.shape
```

```python
data1.columns = [x.replace(" ","_").replace("(","").replace(")","") for x in data1.columns]
data1.head()
```

```python
data1.vehicle_type.value_counts()
```

```python
data1['vehicle_type'] = np.where(data1.vehicle_type=='Xheavy Truck', 'Heavy Truck',
                                 np.where(data1.vehicle_type=='Xheavy Tractor', 'Heavy Tractor',data1.vehicle_type))
data1['vehicle_type'].value_counts()
```

```python
data1.radius.value_counts()
```

```python
data1['radius'] = data1['radius'].str.replace(r'[^0-9.]', '', regex=True)
data1['radius'] = pd.to_numeric(data1['radius'], errors='coerce')
data1['radius'].value_counts()
```

```python
data1['driver_gender'].value_counts()
```

```python
data1['driver_gender'] = (data1['driver_gender']=='M').astype(int)
```

```python
data1.garaging_area.value_counts()
```

```python
data1.groupby('garaging_area')['claim_frequency'].mean()
```

```python
garaging_mapping = {'rural':1, 'suburban':2, 'urban':3}
data1['garaging_area_map'] = data1['garaging_area'].map(garaging_mapping)
```

```python
data1.groupby('vehicle_type')['claim_frequency'].mean()
```

```python
garaging_mapping = {'Trailer':1, 'Heavy Truck':2, 'Heavy Tractor':3}
data1['vehicle_type_map'] = data1['vehicle_type'].map(garaging_mapping)
```

```python
data1.groupby('commodity')['claim_frequency'].mean()
```

```python
commodity_dummies = pd.get_dummies(data1['commodity'], prefix='commodity', drop_first=True, dtype=int)
data1 = pd.concat([data1, commodity_dummies], axis=1)
data1.head()
```

```python
data2 = data1.drop(['vehicle_type','commodity','garaging_area'], axis=1)
data2.head()
```

```python
data2.prior_insurance_limit.describe()
```

```python
data2['prior_insurance_limit_scaled'] = data2['prior_insurance_limit'] / 100000
data2.head()
```

```python
data2.drop('prior_insurance_limit', axis=1, inplace=True)
```

```python
data2.to_csv(r'input_data_w_formatted.csv', index=False)
```
